```python
import pandas as pd

# Load your TSV file (update path as needed)
df = pd.read_excel(r"D:\vr-pipeline\data\Participants.tsv.xlsx", header=2)

# Clean participant IDs: keep only last 6 characters (e.g., EK_401)
df['Studien-ID'] = df['Studien-ID'].str[-6:]
df = pd.DataFrame({
    'participant_id': df['Studien-ID'].tolist() * 2,  # repeat each participant twice
    'session': [1] * len(df) + [2] * len(df),      # session 1 for first set, 2 for second
    'drink': pd.concat([df['Drink intervention 1'], df['Drink intervention 2']], ignore_index=True)
})
df = df.iloc[1:].reset_index(drop=True)
df.replace({'Eris': 'ketone', 'Atlas': 'placebo'}, inplace=True)
df
```

```python
df2=pd.read_csv(r'D:\vr-pipeline\data\results\28052025_segment_data_t123_30s_30s.csv')
merged_df = df2.merge(df, on=['participant_id', 'session'], how='left')
```
