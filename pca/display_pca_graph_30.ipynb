```python
df_30_dark_maze=df_30[df_30['trial'] == 15]
#display(df_90)
df_30_dark_maze_pca = df_30_dark_maze.drop(columns=['participant_id', 'center_dist', 'edge_dist','drink', 'session','trial', 'version', 'segment', 'stops_count','unsafe_area_coverage', 'safe_area_coverage', 'safe_time_spent', 'unsafe_time_spent','outside_time_spent'])

#display(df_90_pca.head())

df_30_pca_x_axis= df_30_dark_maze[['drink','session']]
df_30_pca_x_axis.loc[df_30_pca_x_axis['drink'] == 'placebo', 'drink'] = -0.5
df_30_pca_x_axis.loc[df_30_pca_x_axis['drink'] == 'ketone', 'drink'] = 0.5
scaler=StandardScaler()
df_30_pca_scaled_dark_maze=scaler.fit_transform(df_30_dark_maze_pca)
```

```python
reduced_3_data_30_df_dark_maze = pd.DataFrame(reduced_3_data_30_dark_maze, columns=[f'PC{i+1}' for i in range(reduced_3_data_30_dark_maze.shape[1])])
```

```python
combined_30_dark_maze = pd.concat([
    df_30_dark_maze[['participant_id', 'segment', 'session', 'drink']].reset_index(drop=True),
    reduced_3_data_30_df_dark_maze[['PC1', 'PC2', 'PC3']].reset_index(drop=True)
], axis=1)
```
